```python
df_sheet_name1 = pd.read_excel('ukraine.xlsx', sheet_name='Importers_2020')
print(df_sheet_name1)
```

```python
df2 = df1.rename(columns={df1.columns[1]:'country_code',
                          df1.columns[2]:'importer',
                         df1.columns[3]:'importer_region',
                         df1.columns[6]:'value_for_ukraine',
                         df1.columns[7]:'weight_for_ukraine'})

df2
```

```python
df3 = df2[['country_code','importer','importer_region','value_for_ukraine','weight_for_ukraine']].copy()
df3
```
